```python
filename = 'cities.csv'
fields = ["name", "timeZone_label", "utcOffset", "homepage", "governmentType_label",
          "isPartOf_label", "areaCode", "populationTotal", "elevation",
          "maximumElevation", "minimumElevation", "populationDensity",
          "wgs84_pos#lat", "wgs84_pos#long", "areaLand", "areaMetro", "areaUrban"]
```

Audit the datatypes that can be found in some particular fields in the dataset.

```python
def audit_file(filename, fields):
    fieldtypes = defaultdict(set)
    unique = []
    
    with open(filename) as f:
        reader = csv.DictReader(f, delimiter = ',')
        for i in range(3):
            next(reader)

        for row in reader:
            for item in fields:
                row[item] = row[item].strip()
                if (row[item] == 'Null') or (row[item] == ''):
                    fieldtypes[item].add('NoneType')
                elif row[item].find("{") != -1:
                    fieldtypes[item].add('list')
                else:
                    try:
                        int(row[item])
                        fieldtypes[item].add('int')
                    except ValueError:
                        try:
                            float(row[item])
                            fieldtypes[item].add('float')
                        except ValueError:
                            fieldtypes[item].add('str')
                
            
                    
    return fieldtypes
```

```python
audit_file(filename, fields)
```

```python
def fix_area(area):
    if (area == 'NULL') or (area == ''):
        return None
    else:
        try:
            return float(area)
        except ValueError:
            location = area.find('|')
            area1 = float(area[1:location])
            area2 = float(area[location+1:-1])
            area1_compare = str(float(area[1:location])).replace('.', '').strip('0')
            area2_compare = str(float(area[location+1:-1])).replace('.', '').strip('0')
            if len(area1_compare) >= len(area2_compare):
                return area1
            else:
                return area2
```

```python
def process_file1(filename):
    
    data = []

    with open(filename, "r") as f:
        reader = csv.DictReader(f)

        #skipping the extra metadata
        for i in range(3):
            next(reader)

        # processing file
        for line in reader:
            # calling your function to fix the area value
            if "areaLand" in line:
                line["areaLand"] = fix_area(line["areaLand"])
            data.append(line)

    return data
```

```python
def fix_name(name):

    if (name == 'NULL') or (name == ''):
        return []
    elif '{' in name:
        name = name.strip('{').strip('}').split('|')
        return name
    else:
        return [name]
```
